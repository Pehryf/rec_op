## 2. Modélisation formelle du TSP

### 2.1 Le graphe

On modélise le TSP par un **graphe complet orienté** (ou non orienté si les distances sont symétriques) :

$$G = (V, E)$$

où :
- $V = \{0, 1, \dots, n-1\}$ est l'ensemble des **sommets** (les villes), avec $|V| = n$
- $E = \{(i, j) \mid i \neq j,\ i, j \in V\}$ est l'ensemble des **arêtes** (les routes entre les villes)
- À chaque arête $(i, j)$ est associé un **coût** $c_{ij}$ (distance, temps, etc.)

---

### 2.2 Les variables de décision

On introduit des **variables binaires** $x_{ij}$ :

$$x_{ij} = \begin{cases} 1 & \text{si l'on emprunte l'arc de } i \text{ vers } j \\ 0 & \text{sinon} \end{cases}$$

---

### 2.3 La fonction objectif

On cherche à **minimiser** le coût total du tour :

$$\min \sum_{i \in V} \sum_{\substack{j \in V \\ j \neq i}} c_{ij} \cdot x_{ij}$$
